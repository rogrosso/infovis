# HTML, CSS, SVG, JavaScript and D3

## Mapping Information: Visual Encoding from Color Models to Web-Technology

In visualization we decouple the logic from rendering. The algorithm backend is independent of the rendering pipeline. 

- The Logic: Calculate geometries, positions, animations. In this lecture we use JavaScript, the engine.
- The Rendering: Web-technologies to generate colored images and animations. In InfoVIS we use SVG/Canvas/WebGL embedded 
in a HTML page.
- The Concept: The Engine vs The Paintbrush

In this course, we utilize JavaScript as our engine for data processing, and SVG/HTML as our creative tools. While 
numerous tools (such as Bpkeh, Plotly, Chart.js) automate this process, comprehending the manual mapping of data to the 
DOM empowers you to craft bespoke, high-performance interactive experiences that transcend conventional templates.

## Why Web-Technology

Visual presentations in information visualization are frequently created using web technology due to several reasons.

- Accessible to everyone with internet access.
- It requires no installation and runs on any device with a web browser.
- In the meantime, advanced techniques support and simplify the implementation for design and interaction.
- Examples of visualizations can be found in 
[Explore Observable](https://observablehq.com/explore).

## The Mechanics of Visualization

- Data (data types)
- Geometry
- Attributes: position, size, orientation
- Textures and Colors
- Rendering: in InfovIS we use Web technologies
- One more dimension: Interaction

## Ingredients of a web application

### Low level of abstraction

- HTML
- CSS (cascading style sheets)
- SVG (scalable vector graphics), 2D vector graphics
- canvas: the HTML element for 2D pixel graphics
- WebGL, 3D graphics

### Higher level of abstraction

- JavaScript 
- libraries for building interactive user interfaces and applications, React, Vue.js,  Angular
- D3 (uses vector graphics)

## SVG

Visual presentations in information visualization are commonly based on SVG graphics.

- A well-known javascript library used for Information Visualization is [D3](https://d3js.org/). 
- SVG: beautiful graphics, not limited by pixel resolution, [Ghostscript Tiger](https://upload.wikimedia.org/wikipedia/commons/f/fd/Ghostscript_Tiger.svg)
Not appropriate for high performance; if performance is an issue, use the HTML canvas element or WebGL 
- SVG tutorial: [w3schools: SVG Tutorial](https://www.w3schools.com/graphics/svg_intro.asp) (introductory, low level)

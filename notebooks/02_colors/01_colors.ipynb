# Color Models (Abstract Representation of Colors)

A color model is a mathematical abstraction or coordinate system that defines how colors can be represented using 
numbers. However, it doesn’t have a direct connection to the physical world.

- Logic determines the dimensions of color.
- Examples include:
    - RGB: an additive color model using primary colors (Red, Green, Blue) as the basis.
    - CMY: a subtractive color model using primary colors (Cyan, Magenta, Yellow) as the basis.
    - HSL/HSV: perceptual dimensions that include hues, saturation, and lightness/value.

**The Nuance:** The RGB color (255, 0, 0) represents the abstract definition of pure red. However, the appearance of 
100% red may vary depending on the device. 

Summarizing: A color model is an abstract representation of colors using a tuple of numbers, the three color coordinates.
RGB is an additive color model with three primary colors (coordinates): red, green, and blue. Additive means: 
$\text{white} = 1  \times  \text{red} + 1 \times \text{green} + 1 \times \text{blue}$, that is $\text{white} = (1,1,1)$,
and $\text{black} = (0,0,0)$. The basis in the RGB color model are $\{red, green, blue\}$ .
CMY (or, more generally, CMYk) is a subtractive color model. The primary colors are cyan, magenta, and yellow. It is 
subtractive: $\text{black} = 1 \times cyan + 1 \times magenta + 1  \times yellow$.
There are also the color model HSV and HSL based on human color perception, where the variable h represents the hue (color).

## Color Space (The Physical Implementation)

A color space is a specific, calibrated instance of a color model. It translates those abstract numbers into a gamut—the actual range of colors that a particular device, such as a screen or printer, can produce.

- Logic: It defines the limits of what can actually be seen or reproduced by a divice
- Examples:
    - sRGB: The standard for webpages
    - Adobe RGV: A wider space used by higher quality devices, e.g. for photography
    - CIELAD (L*a*b*): A space designed to be perceptually uniform, meaning the distance between colors reflects their 
    perceived difference to our eyes.

## From Color Model to Color Space

- Selection (Model): You might opt for the HSL model to design your visualization because it’s intuitive. For instance, 
you decide that *Category A* will have a Hue of  $120\circ$  and *Category B* a Hue of $240\circ$.
- Mapping (Space): When you render that visualization, your software maps those HSL values into a specific Color 
Space (like sRGB) so that the user’s screen knows exactly which physical wavelengths of light to emit.
- The “Translation” Problem: If your Color Model (the recipe) specifies a color that falls outside the Gamut of your 
Color Space (the kitchen), the color will be “clipped” or shifted. This is why a bright neon blue on your screen 
(RGB) often appears dull when printed (CMYK).

## RGB Color Model

- primary colors: red, green, and blue
- used in computer graphics programs and monitors
- additive: adding primitive colors produces white
- red = (1,0,0),  green = (0,1,0), blue = (0,0,1)
- black = (0,0,0) = 0*red + 0*green + 0*blue
- white = (1,1,1) = 1*red + 1*green + 1*blue
- cyan = (0,1,1) = 0*red + 1*green + 1*blue

## CMY Color Model

- primary colors: cyan, magenta, and yellow
- used in hard copy devices, e.g. printer
- subtractive: adding primary colors produces black
- black = (1,1,1)
- white = (0,0,0)
- The transformation from RGB:
    - cyan = 1 - red
    - magenta = 1 - green,
    - yellow = 1- blue

## CMYK Color Model

- color printers commonly use this model
- modification of CMY-model
- K for black, addition (redundant) color to save ink
- no unique representation: (1,1,1,0) and (0,0,0,1) for black
- Standard transformation from CMY to CMY Kkvar = min(C,M,Y), 
C = (C - kvar) / (1 - kvar), M = (M - kvar) / (1 - kvar), Y = (Y - kvar) / (1 - kvar), K = kvar

## How to Transform from RGB to HSV

Three parameters: H: hue, S: saturation, V: largest component of color, where 
$H \in [0, 360]$, and $S, L \in [0,1]$; $\($R, G, B \in [0,1]$

$$
\begin{align*}
M &= \max(R,G,B), \, m = \min(R,G,B)\\
C & = M - m\\
H' &= 
\begin{cases}  \text{undefined if } C =0\\
   \frac{G-B}{C}\mod 6, \text{if } M=R\\
   \frac{B-R}{C} + 2, \text{if } M=G\\ 
   \frac{R-G}{C} + 4, \text{if } M=B\\ 
\end{cases}\\                                
H &= 60 \times H', \text{ if } H < 0, \text{ then } H = H + 360\\ 
V &= M\\
S &= 
\begin{cases}
0, \quad \text{if } V=0\\
\frac{C}{V}, \quad \text{otherwise}
\end{cases}
\end{align*}
$$

## The HSL Color Model

Three parameters: H: hue, S: saturation, L: lightness (relative brightness), where 
$H \in [0, 360]$, and $\($S, L \in [0,1]$\)$; $\(R, G, B \in [0,1]$
$$
\begin{align*}
 M &= \max(R,G,B), \, m = \min(R,G,B)\\
C &= M - m\\
H' &= 
\begin{cases}
\text{undefined if } C =0\\
\frac{G-B}{C}\mod 6, \text{if } M=R\\
\frac{B-R}{C} + 2, \text{if } M=G\\
\frac{R-G}{C} + 4, \text{if } M=B\\
\end{cases}\\
H &= 60 \times H'\\ 
L &= \frac{1}{2}(M+m)\\
S &= 
\begin{cases}
0, \text{if } L=1 \lor L=0\\
\frac{C}{1-|2L - 1|}, \text{otherwise}
\end{cases}
\end{align*}
$$

## The HCL Color Model

The HCL color model is a polar transformation of the CIELAB color space. It is designed to be perceptually 
uniform, ensuring that equal numerical changes in the model correspond to equal perceptual changes in color.

- Hue (\(H\)): The type of color (e.g., Red, Green, Blue) expressed as an angle from 
$0^\circ  \text{to} 360^\circ$.
- Chroma ($C$): The purity or "colorfulness." Unlike Saturation, Chroma is absolute; it measures how 
much a color differs from a gray of the same luminance.
- Luminance ($L$), the perceived brightness, is the most crucial dimension because it closely aligns with 
how the human eye perceives light intensity.

### The Space vs. Model Confusion

There is significant ambiguity in literature regarding whether HCL is a "model" or a "space."

- The Nuance: Technically, HCL is a color model (a coordinate system) that is built on top of the CIELAB color space.
- HCL is often confused because it is mathematically linked to the CIE standards of human vision, which 
are device-independent. This makes it a Color Space. Unlike RGB or HSV, which vary depending on the monitor, 
HCL represents an absolute physical reality of human sight.

### HCL vs. HSV

At first glance, HCL and HSV appear to be siblings because they both employ a “color wheel” logic. 
However, their functions are quite distinct.

## Web Colors / HTML Color Codes

Web colors are just colors. They are usually displayed using an RGB- or HSL-triplet or a hexadecimal 
format (hex-code or -triplet)

#### Format hex-triplet
HTML color code consists of a # and three integer numbers in hex format and is represented using strings.

#### Colors
the three numbers are the RGB components of the color in the range [0,255]. 
It uses one byte for each color.

#### Examples
red: [255,0,0] in RGB, the HTML code is #FF0000
cyan: [0,255,255] in RGB, the HTML code is #00FFFF

CSS Color Keywords

These are case-insensitive identifiers that represent a defined (in terms of RGB or hex-triplet value) color, 
such as magenta.

The list of the accepted color keywords up to CSS level 4 can be found at the MDN under 
[CSS: Cascading Style Sheet](https://developer.mozilla.org/en-US/docs/Web/CSS/color_value)

## Color Perception

When we discuss color perception, we transition from physics (the wavelengths of light) to psychology 
(how our brains interpret those wavelengths). The HSV (Hue, Saturation, Value) model was specifically 
designed to better align with human experience than the machine-oriented RGB model.

### The Perceptual Dimensions of HSV

Humans don’t naturally think in terms of “how much Red or Green is in this color.” Instead, we perceive 
three distinct qualities that the HSV model directly maps: hue, saturation, and value.

- Hue ( 𝐻 ): This answer the question "What color is it?" ( 0∘→360∘ ). It corresponds to the dominant wavelength. 
- Saturation ( 𝑆 ): This is the "purity" or "intensity" of the color ( 0→1 ). A high saturation is vivid; 
a low saturation looks grayish.
- Value ( 𝑉 ): Also called Brightness ( 0→1 ). This is how much light appears to be reflected. 

### Intuition vs. Reality

The primary reason HSV is popular in InfoVis is its flexibility in working with different color models. 
In RGB, you have to mathematically decrease  𝑅 ,  𝐺 , and  𝐵  simultaneously to make a color “darker.” 
However, in HSV, you simply slide the Value ( 𝑉 ) down to achieve the desired effect. Additionally, 
HSV allows you to focus on setting the “color” and then consider value and saturation, which are 
clearly human perceptions. 

However, there is a famous catch in how we perceive these: Perceptual Non-Uniformity. The Problem: Even 
if two colors have the same "Value" ( 𝑉 ) in the HSV model, they might not look equally bright to a human. 
For example, a pure Yellow ( 𝐻=60,𝑉=1 ) looks much brighter to our eyes than a pure" Blue ( 𝐻=240,𝑉=1 ).

### Why this Matters

Due to the way our eyes perceive colors, the HSV model can sometimes be misleading. While HSV is an 
improvement over RGB for designing visualizations (since it aligns with our color vocabulary), it remains 
a “geometric” model, not a truly “perceptual” one. For high-accuracy scientific visualization, we often 
switch to HCL (Hue, Chroma, Luminance), which mathematically corrects for the fact that we perceive Yellow 
as brighter than Blue.

```python
from typing import cast
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_swiss_roll, make_s_curve, make_blobs
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import umap

# -----------------------------
# Helper: run all embeddings
# -----------------------------
def compute_embeddings(X):
    results = {}

    # PCA
    pca = PCA(n_components=2)
    results["PCA"] = pca.fit_transform(X)

    # t-SNE
    tsne = TSNE(n_components=2, perplexity=30, random_state=42, init="pca")
    results["t-SNE"] = tsne.fit_transform(X)

    # UMAP
    umap_model = umap.UMAP(
        n_components=2,
        n_neighbors=15,
        min_dist=0.1,
        random_state=42
    )
    results["UMAP"] = umap_model.fit_transform(X)

    return results


# -----------------------------
# Helper: plot results
# -----------------------------
def plot_results(X, color, title):
    embeddings = compute_embeddings(X)

    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    methods = ["PCA", "t-SNE", "UMAP"]

    for ax, method in zip(axes, methods):
        Y = embeddings[method]
        ax.scatter(Y[:, 0], Y[:, 1], c=color, s=5, cmap="Spectral")
        ax.set_title(method)
        ax.set_xticks([])
        ax.set_yticks([])

    fig.suptitle(title, fontsize=14)
    plt.tight_layout()
    plt.show()


# =============================
# (A) Swiss Roll
# =============================
X, color = make_swiss_roll(n_samples=2000, noise=0.05)
plot_results(X, color, "Swiss Roll")


# =============================
# (B) S-Curve
# =============================
X, color = make_s_curve(n_samples=2000, noise=0.05)
plot_results(X, color, "S-Curve")


# =============================
# (C) Well-separated clusters
# =============================
result = make_blobs(
    n_samples=2000,
    centers=4,
    cluster_std=0.5,
    random_state=42,
    return_centers=False,
)
X, labels = cast(tuple[np.ndarray, np.ndarray], result)
plot_results(X, labels, "Well-separated clusters")

# =============================
# (D) Varying density clusters
# =============================
result = make_blobs(
    n_samples=2000,
    centers=4,
    cluster_std=[0.2, 0.5, 1.5, 3.0],
    random_state=42,
    return_centers=False,
)
X, labels = cast(tuple[np.ndarray, np.ndarray], result)
plot_results(X, labels, "Varying density clusters")
```

## Histograms
We learn to visualize data using histograms in Python. A histogram represents the distribution of discrete quantitative data. This visualization will illustrate the central limit theorem through sequences of samples generated from a normal distribution.

###  Example: Central Limit Theorem
We estimate (infer) the sample mean and the sample standard deviation from a large population using the mean of sample statistics. 

### Law of Large Numbers
Given a sequence of i.i.d. random variables 
$$
X_i,\; i =1,\dots,n
$$
with mean $\mu$ and standard deviation $\sigma$, then the sample mean converges to $\mu$ as $n$ tends to $\infty$, that is
$$
\bar{\mu} = \frac{1}{n} \sum X_i \; \rightarrow \mu \; \text{for } n \rightarrow \infty
$$

### Central Limit Theorem
For large $n$ the sample mean is normaly distributed with mean $\mu$ and standard deviation $\frac{\sigma}{\sqrt{n}}, that is
$$
\bar{mu} = \frac{1}{n} \sum X_i \sim \mathcal{N}(\mu, \frac{\sigma^2}{n})
$$

### Sampling (survey sampling)
From a large population with $N$ samples, we sample smaller populations of size $n \ll N$ and compute the sample means and standard deviations distribution. 

The large population with $\mu$ and $\sigma$
$$
\{x_1, \dots , x_N\}\, , \; 
\mu = \frac{1}{N} \sum xi \, , \; 
\sigma^2 = \frac{1}{N} \sum (x_i - \mu)^2
$$
The sample statistics is given by
$$
\{X_1,\dots, X_n\}\, , \; 
\bar{\mu} = \frac{1}{n} \sum X_i \, ,\; 
\bar{\sigma} = \frac{1}{n} \sum (X_i - \bar{\mu})^2
$$

We compute a large number of sample statistics from a large population and visualize the distribution of the sample means. In this example we will plot the relative frequencies in order to compare frequencies with the probability density function of a normal distribution.

```python
# generate an array of normally distributed random numbers
# survey sampling: compute the distribution of the sample mean of a set of smaller populations 
# from a larger population
# the sample mean is normally distributed
# the variance of the sample mean is the variance of the population divided by the number of samples
import numpy as np
import matplotlib.pyplot as plt

# set seed for reproducibility
np.random.seed(11) # like prime numbers, 11 is a good seed

# mean and standard deviation of the large population
mu = 100
sigma = 15

# generate the large population
N = 100000
population = np.random.normal(mu, sigma, N)

# number of smaller populations for sample survey sampling statistics
m = 1000
# number of samples in each smaller population
n = 300

# generate the sample means
sample_means = np.zeros(m)
for i in range(m):
    sample = np.random.choice(population, n)
    sample_means[i] = np.mean(sample)

# mean and standard deviation of the sample mean
mu_sample = np.mean(sample_means)
sigma_sample = np.std(sample_means)

print('Mean of the population:', mu)
print('Standard deviation of the population (divideb bei sqrt(n):', sigma/np.sqrt(n))
print('Mean of the sample mean:', mu_sample)
print('Standard deviation of the sample mean:', sigma_sample)

# plot the sample means
plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.7, density=True)
plt.xlabel('Sample mean')
plt.ylabel('Density')
plt.title('Distribution of the sample mean')

# plot the normal distribution with the same mean and standard deviation as the sample mean
x = np.linspace(mu_sample - 4*sigma_sample, mu_sample + 4*sigma_sample, 100)
y = 1/(sigma_sample * np.sqrt(2*np.pi)) * np.exp(-(x - mu_sample)**2 / (2*sigma_sample**2))
plt.plot(x, y, color='red', lw=2)
#plt.savefig("histogram.png")    
plt.show()
```

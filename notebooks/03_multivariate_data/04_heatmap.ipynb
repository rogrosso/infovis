# Heatmap matrix
In a heatmap matrix, the rows are the observations in a multivariate data set, and the columns are the attributes measured. The attributes might be categorical or quantitative data. The values of these attributes are color-coded and represented as matrix entries; see the plot below. A heatmap matrix might reveal a pattern in the data by showing if the values of different attributes are similar, thus detecting correlations, or it might visualize outliers if one entry strongly differs in color.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read data
#demographic = pd.read_csv('../data/INED-DemographicData.csv', sep=';')
demographic = pd.read_json('../../data/INED-DemographicData.json', orient='records')
demographic.head()
# convert 
```

### Plot heatmap with Matplotlib

```python
eu_demo = demographic[demographic['Continent'] == 'Europe']
cols = ['Country', 'Pop', 'Birth rate', 'Mortality rate', 'Life expectancy', 'Infant mortality', 'Children per woman', 'Growth rate', 'Population aged 65+']
eu_demodata = eu_demo[cols]
eu_sorted = eu_demodata.sort_values(by='Pop', ascending=False).head(10)
countries = eu_sorted['Country']

# Normalize data for better visualization
maxv = eu_sorted.iloc[:,1:].max().max()
eu_copy = eu_sorted.copy()
# columne wise normalization
for col in cols[1:]:
    col_max = eu_sorted[col].max()
    eu_sorted[col] = eu_sorted[col] / col_max
#eu_sorted.style.background_gradient(cmap='viridis')

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
#ax.imshow(eu_sorted[cols[1:]], cmap='coolwarm', interpolation='nearest', aspect='auto')
ax.imshow(eu_sorted[cols[1:]], cmap='twilight', interpolation='nearest', aspect='auto')
# Show all ticks and label them with the respective list entries
ax.set_xticks(range(len(cols)-1), labels=cols[1:], rotation=45, ha="right", rotation_mode="anchor")
ax.set_yticks(range(len(countries)), labels=countries)

# Loop over data dimensions and create text annotations.
for i in range(len(countries)):
    for j in range(len(cols)-1):
        text = ax.text(j, i, eu_copy.iloc[i, 1+j],ha="center", va="center", color="w")

ax.set_title("Demographic data for the 10 most populous European countries (pop in millions)")
fig.tight_layout()
plt.show()
```

### Heatmaps with seaborn

```python
eu_sns = eu_sorted.set_index('Country')
fig, ax = plt.subplots(figsize=(10, 6))
p = sns.heatmap(eu_sns, cmap='coolwarm', annot=True, fmt=".2f", cbar=False).set_title("Demographic data for the 10 most populous European countries (pop in millions)") 
ax.tick_params(axis='x', labelrotation=50)
plt.show()
```

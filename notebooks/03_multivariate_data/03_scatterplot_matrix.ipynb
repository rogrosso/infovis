# Scatter Plot Matrix
A scatter plot matrix, also known as a pair plot, is a grid of scatter plots that visualizes the relationships between multiple variables in a multivariate dataset. Each variable is plotted against every other variable in the matrix, making it a powerful tool for exploring multivariate data.
### Properties
- generalization of the scatter plot
- represent a large number of multivariate attributes simultaneously
- visualize all possible pairs of attributes with a scatter plot
- order the scatter plots in a 2D array
- correlations in the data become easily visible
- the diagonal can be used to visualize a histogram for the corresponding variable or any other statistically relevant information such as the kernel density estimate (kde), etc.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load data data 
cars = pd.read_json('../../data/cars.json')
for key in cars.keys():
    if np.issubdtype(cars[key].dtype, np.number):
        cars[key] = pd.to_numeric(cars[key], errors='coerce')
print(cars.head())
```

### Scatter plot

```python
fig, ax = plt.subplots()
# marker size
msize = 150 * cars['displacement'] / cars['displacement'].max()
#print(msize)
print(len(msize))
print(len(cars['mpg']))
print(len(cars['weight']))
# plot
sc = ax.scatter(cars['mpg'], cars['weight'], c=cars['cylinders'], s=msize, alpha = 0.5)
# labels
ax.set_xlabel('mpg')
ax.set_ylabel('weight')
# colorbar
cbar = fig.colorbar(sc)
cbar.set_label('cylinders')
#plt.savefig("scatterplot.png")
plt.show()
```

### Scatter plot matrix
In this example, we use the visualization functionality from Pandas, which is essentially a wrapper around the well-known and widespread Python visualization toolkit matplotlib. We arrange the plots in a matrix to visualize many variables with scatter plots. The diagonal in the plot below shows the variables' Kernel Density Estimation (an estimate of the probability density function). 

```python
cols = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
p = pd.plotting.scatter_matrix(cars[cols], alpha=0.2, figsize=(10, 10), diagonal='kde')
```

### Seaborn
In the example below, we use the Seaborn visualization package to visualize numeric data with a scatter plot matrix. According to the documentation, the diagonal represents the marginal distribution of the data in each column. The Seaborn visualization package is recommended for applications in statistical data science.

```python
sns.set_theme(style="ticks")
# pairplot
cols = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration', 'cylinders']
# use number of cylinders to set color
sns.pairplot(cars[cols], hue='cylinders', palette='Accent')
#plt.savefig("scatterplot_matrix.png")
plt.show()
```

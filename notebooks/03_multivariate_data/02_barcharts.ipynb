# Bar Charts
Bar charts (or bar graphs or plots) efficiently represent categorical data, using vertical or horizontal rectangular bars to visualize the frequency of different items within a category. This visualization technique uses rectangular bars, which can be vertically or horizontally oriented.

### Functioning of bar charts:
- Categorical data: Each bar represents the proportional amount of a specific entity in the total number of individuals in the category (e.g., the total number of apples proportional to the total number of fruits or amount of persons with a certain education level in relation to the population).

- Geometry: Bars are rectangular, and the height for a vertical plot or the length for the horizontal case represents the count of the corresponding entity in the category. 
 
- Axes: In the horizontal plot, the ticks on the x-axis represent the entities in the category, and the y-axes give the count of each entity in the total number of category members. In the vertical plot, the x- and y-axes change roles. 

- Visual presentation: Bars are usually slightly separated and have a darker edge to emphasize the different entities in the category. Colors can be used to make the visual presentation more appealing. Adding labels will make the presentation more readable. 

Bar charts are particularly helpful in visualizing proportionality across the different entities in a category.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load data data 
titanic = pd.read_csv('../../data/titanic.csv')
titanic.head()
```

### Plot barcharts using pandas
The popular data analysis tool ***pandas*** uses the method *plot()* with different arguments to visualize data. To visualize the graphic on screen one uses the pyplot submodule of Matplotlib.

```python
# barplot survival rate by class and sex
titanic.groupby(['Pclass','Sex','Survived'])['Survived'].count().unstack().plot(kind='bar')
plt.title('Bar plot of survival rate by class and Gender')
plt.xlabel
#plt.savefig("barplot_survival_rate.png")
plt.show()
```

### Plot barcharts using Matplotlib
We can use Matplotlib for plotting categorical variables. 

```python

def get_column_count(data_frame, pclass, gender, survived):
    return titanic \
        .loc[(titanic['Pclass'] == pclass) & (titanic['Sex'] == gender) & (titanic['Survived'] == survived)] \
        .groupby(['Pclass','Sex'])['Survived'].count()

female_survived = {
    'survived': np.array([get_column_count(titanic, 1, 'female', 1), get_column_count(titanic, 2, 'female', 1), get_column_count(titanic, 3, 'female', 1)]).flatten(),
    'perished': np.array([get_column_count(titanic, 1, 'female', 0), get_column_count(titanic, 2, 'female', 0), get_column_count(titanic, 3, 'female', 0)]).flatten(),
}


fig, ax = plt.subplots()
bottom = np.zeros(3)
width = 0.5
pclass = ('1st class', '2nd class', '3rd class')
for key, value in female_survived.items():
    p = ax.bar(pclass, value, width, label=key, bottom=bottom)
    bottom += value
    ax.bar_label(p, label_type='center')

ax.legend()
plt.title('Bar plot of survival rate by class')
plt.xlabel('Class and survival status')
plt.ylabel('Count')
plt.show()
```

### Plot barchart with given colors

```python
def get_cc(data_frame, pclass, gender, survived):
    res = titanic \
        .loc[(titanic['Pclass'] == pclass) & (titanic['Sex'] == gender) & (titanic['Survived'] == survived)].groupby(['Sex'])['Sex'].count()
    return res.iloc[0]

# plot bars with colors
bar_colors = ['#ff8000', '#bfff00']
survived = [get_cc(titanic,1,'female',1),get_cc(titanic,2,'female',1),get_cc(titanic,3,'female',1)]
perished = [get_cc(titanic,1,'female',0),get_cc(titanic,2,'female',0),get_cc(titanic,3,'female',0)]
pclass = ['1st class', '2nd class', '3rd class']
fig, ax = plt.subplots()
bar_width = 0.3
index = np.arange(len(pclass))
bar1 = ax.bar(index, survived, bar_width, label='Survived', color=bar_colors[0])
bar2 = ax.bar(index + bar_width, perished, bar_width, label='Perished', color=bar_colors[1])
ax.set_xlabel('Class')
ax.set_ylabel('Count')
ax.set_title('Bar plot of survival rate by class for female passengers')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(pclass)
ax.legend()
ax.bar_label(bar1, label_type='center')
ax.bar_label(bar2, label_type='center')
plt.show()

```

### Plot data using Matplotlib pies

```python
survived = [[get_cc(titanic,1,'female',1),get_cc(titanic,1,'female',0)], \
            [get_cc(titanic,2,'female',1),get_cc(titanic,2,'female',0)], \
            [get_cc(titanic,3,'female',1),get_cc(titanic,3,'female',0)]]
fig, ax = plt.subplots()
size = 0.3
vals = np.array(survived)
tab20c = plt.color_sequences['tab20c']
colors = [
    '1st class, total',
    '2nd class, total',
    '3rd class, total',
    '1st class, survived',
    '1st class, perished',
    '2nd class, survived',
    '2nd class, perished',
    '3rd class, survived',
    '3rd class, perished',
]
outer_colors = [tab20c[0], tab20c[4], tab20c[8]]
inner_colors = [tab20c[1], tab20c[2], tab20c[5], tab20c[6], tab20c[9], tab20c[10]]
ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors, wedgeprops=dict(width=size, edgecolor='w'))
ax.pie(vals.flatten(), radius=1-size, colors=inner_colors, wedgeprops=dict(width=size, edgecolor='w'))
ax.set(aspect="equal", 
       title='Pie plot of survival rate by ticket class for female passengers',
       )
ax.legend(colors,
          loc='upper right',
          bbox_to_anchor=(1.3, 0, 0.5, 1),
          fontsize='small',
          fancybox=True,
          shadow=True)
plt.setp(ax.patches, width=0.3, edgecolor='w')
plt.show()
```

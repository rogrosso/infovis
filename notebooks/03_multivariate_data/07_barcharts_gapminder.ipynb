# Case Analysis
## Bar Charts
Bar charts are a widely used visualization technique for displaying categorical data. One axis represents the categories, while the other represents a continuous variable, illustrating quantity or count. The quantities within each category are represented by rectangles, or bars, with the size of each bar reflecting the number of items in that category.

### Data: Population data from Gapminder
[Gapminder](https://www.gapminder.org/) is a non-profit organization founded by Hans Rosling, Anna Rosling Rönnlund, and Ola Rosling. It promotes a fact-based worldview by educating people about global development through data visualization. The organization is well-known for its interactive visual tools that provide insight into global trends in health, economics, education, and the environment. In this case study, we visualize data using different bar charts.

### Vertical Bars

```python
from bokeh.sampledata.gapminder import fertility as fer
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.io import output_notebook, export_png
from bokeh.palettes import Bright5, HighContrast3
from bokeh.transform import factor_cmap
from bokeh.io import export_png

output_notebook()

fd = fer.loc[['Argentina', 'Brazil', 'Chile', 'Colombia', 'Peru'], 
             ['1964', '1984', '2013']]
#fd.head()

x_labels = list(fd.index.array)

p = figure(title="Women Fertility in South America",
            x_range=x_labels, 
            height=350,
            toolbar_location=None,
            tools="hover",
            tooltips="@x: @top",)

pval = p.vbar(x=x_labels, 
       top=fd['1964'].values, 
       width=0.9,
       legend_label="1964",
       alpha=0.8)

show(p)
```

### Colors

```python
source = ColumnDataSource(data=dict(
    x=x_labels,
    y=fd['1964'].values,
    color=Bright5,
    label=["1964"] * len(x_labels)
))

p = figure(title="Women Fertility in South America",
            x_range=x_labels, 
            y_range=(0, 10),
            height=350,
            toolbar_location=None,
            tools="hover",
            tooltips="@x: @y",)
p.vbar(x='x', top='y', width=0.9, source=source,
       legend_field='x',
       fill_color='color',
       alpha=0.8)



p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"
p.legend.label_text_font_size = "8pt"
show(p)
```

### Horizontal Bars

```python
y_labels = list(fd.index.array)
source = ColumnDataSource(data=dict(
    y=y_labels,
    x=fd['1964'].values,
    color=Bright5,
    label=["1964"] * len(x_labels)
))

p = figure(title="Women Fertility in South America",
            x_range=(0,10), 
            y_range=y_labels,
            height=350,
            toolbar_location=None,
            tools="")
p.hbar(y='y', right='x', width=0.9, source=source,
       height=0.9,
       legend_field='y',
       fill_color='color',
       alpha=0.8)

p.ygrid.grid_line_color = None
p.legend.orientation = "vertical"
p.legend.location = "center_right"
p.legend.label_text_font_size = "8pt"
#export_png(p, filename="bokeh_hbar.png")
show(p)
```

### Stacking

```python
source = {
    'country': x_labels,
    '1964': fd['1964'].values,
    '1984': fd['1984'].values,
    '2013': fd['2013'].values,
}

p = figure(title="Women Fertility in South America",
            x_range=x_labels, 
            y_range=(0, 20),
            height=350,
            toolbar_location=None,
            tools="hover",
            tooltips="$name: @country, @$name")
years = list(fd.columns)

p.vbar_stack(
            years,
            x='country',
            width=0.9,
            source=source,
            color=HighContrast3,
            legend_label=years,
        )
p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"
#export_png(p, filename="bokeh_vbar_stack.png")
show(p)
```

### Grouping

```python
countries = list(fd.index.array)
years = list(fd.columns)
x = [ (country, year) for country in countries for year in years ]
fertility = [ fd.loc[country, year] for country in countries for year in years ]

source = ColumnDataSource(data=dict(x=x, fertility=fertility))

p = figure(
    title="Women Fertility in South America",
    x_range=FactorRange(*x),
    height=350,
    toolbar_location=None,
    tools="hover",
    tooltips="@x: @fertility",
)

p.vbar(x='x', top='fertility',
        width=0.9, 
        source=source,
        fill_color=factor_cmap('x', palette=Bright5, factors=years, start=1, end=2)
        )
p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None
#export_png(p, filename="bokeh_vbar_group.png")
show(p)
```

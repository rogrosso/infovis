# Visualization of the Population of American Countries
In this example, we visualize the population of some of the largest countries in America using a pie chart. A pie chart is a circular graphic divided into slices to proportionate quantities. Each slice presents the percentage of one of the classes in the data. We are using the sampledata module from the Python visualization package Bokeh. The data is collected by the independent Swedish foundation Gapminder. 

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import Legend
from bokeh.palettes import Spectral6
from bokeh.sampledata.gapminder import population, regions
from bokeh.palettes import Category20c
from bokeh.transform import cumsum
from math import pi
import pandas as pd

# stay in notebook mode
output_notebook()

# get list of countries in America
america = regions[regions['Group'] == 'America'].copy()
list_america_countries = america.index.tolist()
# get countries in America with more than 10M people
population_america = population.loc[america.index]
population_2013 = population_america.loc[population_america['2013'] > 10000000, '2013']
# population in millions
population_2013 = population_2013 / 1000000
population_2013 = population_2013.round(0).astype(int)
#print('population_america: ', population_2013)

# create a new dataframe with the population of the countries in America
# add colors and angles for the pie chart
pp = pd.DataFrame(population_2013.rename('population'))
pp['color'] = Category20c[len(pp)]
pp['angle'] = pp['population']/pp['population'].sum() * 2*pi
#print('new name population_2013: ', pp.head())

# create figure and set title and tools
p = figure(
    title="Gapminder Data: 2013 Population of largest American Countries",
    width=700, height=700,
    tooltips='@Country: @population M',
    toolbar_location="below")
p.add_layout(Legend(),'right')
# create pie chart, use a wedge glyph and cumsum transform to calculate the angle of each wedge
p.wedge(
    x=0, y=1, radius=0.45,
    start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
    line_color="white", fill_color='color', 
    #legend_field='Country', 
    legend_group='Country',
    source=pp
    )
p.axis.axis_label = None
p.axis.visible = False
p.grid.grid_line_color = None
#leg = p.legend[0]
show(p)
```
